### **Terms in Perceptron**

**Input Values (x1, x2, ..., xn)**: These are the numerical values that the perceptron receives as input. Each input represents a feature or attribute of the data being processed.

**Weights (w1, w2, ..., wn)**: Each input value is associated with a weight, which determines the strength or importance of that input in influencing the perceptron's output. Weights are adjusted during the learning process to optimize the perceptron's performance.

**Bias (b)**: The bias is a constant value added to the weighted sum of inputs. It allows the perceptron to shift the activation function, enabling it to model more complex relationships between inputs and outputs.

**Weighted Sum ()**: The weighted sum is calculated by multiplying each input value by its corresponding weight and summing the results, then adding the bias term:  = (w1 * x1) + (w2 * x2) + ... + (wn * xn) + b.

**Activation Function (f)**: The activation function introduces non-linearity into the perceptron's output. It takes the weighted sum as input and produces a final output value. Common activation functions include the step function, sigmoid function, and ReLU function.

**Output (y)**: The output of the perceptron is the result of applying the activation function to the weighted sum: y = f(). The output represents the perceptron's decision or prediction based on the given input values.

#### Learning Rate

The perceptron learning rule is a way to adjust the weights and bias of a perceptron (a single-layer neural network) to make it learn from data.

Here's how it works:

1. Initialization: Start by initializing the weights and bias with random values.

2. Prediction: For each input data point, calculate the weighted sum of the inputs, add the bias, and apply an activation function (usually a step function) to get the predicted output.

3. Error Calculation: Compare the predicted output to the actual output (ground truth) and calculate the error.

4. Weight Update: Adjust the weights and bias based on the error and the input values. The amount of adjustment is controlled by a learning rate.

5. Repeat: Repeat steps 2-4 for all data points in the training set (an epoch). Continue training for multiple epochs until the error is minimized.

Key Points:

The perceptron learning rule is an iterative algorithm.
It only updates the weights when the perceptron makes a mistake (misclassifies a data point).
The learning rate determines how much the weights are adjusted in each update.
The perceptron learning rule is guaranteed to converge if the data is linearly separable.
