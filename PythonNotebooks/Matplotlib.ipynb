# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.Co-ordinates

```python
plt.plot([2,4,6,8,10],"b")
plt.xlabel("Names")
plt.ylabel("grades")
plt.show()
```

## 2.Grid

```python
plt.plot([2,4,6,8],[1,3,5,7],"bo")
plt.grid()
plt.show()
```

## 3.Bargraph

```python
x=np.array(["a","b","c","d"])
y=np.array([10,20,50,30])
plt.bar(x,y)
plt.show()
```

## 4.Pie

```python
x=["may","june","april","july"]
y=np.array([12,24,18,36])
plt.pie(y,labels=x)
plt.show()
```
