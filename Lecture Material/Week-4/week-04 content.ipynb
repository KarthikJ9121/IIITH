# Computing Minimum Edit Distance using Dynamic Programming

- consider two strings X,Y of length n,m respectively.
- we define D(i,j)
    - the edit distance between X[1..i] and Y[1..j]
    - i.e.,the first i characters of X and the first j characters of Y
- The edit distance between X and Y is thus D(n,m)


#### - To compute minimum edit distance we use the concept of Dynamic Programming
- Dynamic programming:A tabular computation of D(n,m)
- It is a Bottom-Up approach:
    - we compute D(i,j) for smaller i,j
    - and larger D(i,j) based on previously computed smaller values
    - i.e compute D(i,j) for all i(0<i<n) and for all j(0<j<m)

# Named Entity Recognition (NER)

### what is named entity recognition ?
Named Entity Recognition is a task of detecting and classifying all the proper names or named entity mentions in a text.

This process of named entity recognition refers to the combined task of finding spans of text that constitute proper names and then classifying the entities according to their type.

In general,NER systems focus on detecting entites like Places,People,Organizations,Date,Time,Numbers,Quantity,Currency etc.

Types of entities to detect is application specific.for example,if your point of interest is to detect bilogical entities like genes,proteins.in such cases, application specific NER systems come into picture.

# Parts of Speech Tagging

#### Part-of-speech tagging is the process of assigning a part-of-speech or other syntactic class marker to each word in a corpus.

Because tags are generally also applied to punctuation, tagging requires that the punctuation marks (period, comma, etc) be separated off of the words.

Thus tokenization is usually performed before, or as part of, the tagging process, separating commas, quotation marks, etc., from words, and disambiguating end-of-sentence punctuation (perod,question mark,etc) from part-of-word punctuation (such as in abbreviation like e.g. and etc.)



parts of speech can be broadly divided into 2 categories.
#### 1.Closed class type:
closed classes are those that has relatively fixed membership. Very rarely new members will be added.

example:prepositions(on, under, over, near, by, at, from, to, with) and determiners(a, an, the)



#### 2.Open class type : 
four major open class types are nouns,verbs,adjectives and adverbs because they are added or borrowed from other languages. 

Here the memborship is not fixed.


#### applications of POS tagging : Text to Speech

#### Example:

```python
import nltk
from nltk.tokenize import word_tokenize
text = word_tokenize("They refuse to permit us to obtain the refuse permit")
nltk.pos_tag(text)
```

#### Ambiguities

For example consider the word Book

it has more than one possible usage and part-of-speech.

it can be a verb in the sentences "book that flight", "book the suspect"

it can be a noun in the sentence "hand me that book"

The problem of POS-tagging is to resolve these ambiguities, choosing the proper tag for the context.

# chunking

a group of words make up phrases.

there are five major phrases they are : 1.Noun Phrase , 2.Verb Phrase , 3.Adjective Phrase , 4.Adverb Phrase, 5.Prepositional Phrase

Chunking is a process of extracting phrases from unstructured text.

It works on top of POS tagging. It uses POS-tags as input and provides chunks as output.

Chunking can break sentences into phrases that are more useful than individual words and yield meaningful results.

Chunking is very important when you want to extract information from text such as locations, person names. (entity extraction)

#### example:
input sentence: The morning flight from Denver has arrived.

output of chunking:
[NP The morning flight] [PP from] [NP Denver] [VP has arrived.]

This bracketing notation makes clear the two fundamental tasks that are involved in chunking: finding the non-overlapping extents of the chunks, and assigning the correct label to the discovered chunks.
